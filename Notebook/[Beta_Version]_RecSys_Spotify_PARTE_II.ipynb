```python
artists = df_artists_resulted['artist_tags']
```

```python
artists
```

```python
tokens = [[tag for tag in track.split()] for track in tracks]
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(tokens)]
```

```python
model = gensim.models.doc2vec.Doc2Vec(min_count=1)
```

```python
model.build_vocab(documents)
```

```python
len(model.wv.vocab)
```

```python
model.train(documents, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model
```

```python
indexer = AnnoyIndexer(model, 10)
```

## Testes

### Procurando a similaridade

```python
def describe_similars(id):
  for i in id:
    track = df_tracks_resulted.iloc[i[0]]
    print("%s by %s, similarity %f" % (track['name'], track['artists'], round((i[1]*100), 2)))
```

```python
similar = model.wv.most_similar(['Alphaville'], topn=15, indexer=indexer)
```

```python
describe_similars(similar)
```

*   As maiores similaridades estão por volta 0.6

### Calculando a similaridade entre 2 vetores semelhantes (2 faixas de um mesmo artista)

```python
df_tracks_resulted[df_tracks_resulted['name'] == 'The Voyager']
```

```python
df_tracks_resulted[df_tracks_resulted['name'] == 'Big in Japan']
```

```python
the_voyager = df_tracks_resulted.iloc[138682]['textual_tags'].split()
```

```python
big_in_japan = df_tracks_resulted.iloc[20316]['textual_tags'].split()
```

```python
v1 = model.infer_vector(the_voyager)
```

```python
v1
```

```python
v2 = model.infer_vector(big_in_japan)
```

```python
v2
```

```python
1 - spatial.distance.cosine(v1, v2)
```

*   Ambas as músicas são do mesmo artista e, portanto, possuem praticamente as mesmas tags (exceto o 'ano')
*   Porém, ao gerar um vetor pra cada uma delas, elas tem baixíssima similaridade
